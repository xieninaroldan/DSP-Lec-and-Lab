# Laboratory Exercise 6
Linear and Non-Linear Systems

```python
import numpy as np
import matplotlib.pyplot as plt
```

**Problem 1: Linear Time-Invariant (LTI) System Response**

```python
N = 100  
n = np.arange(N)  

x = np.sin(0.2 * np.pi * n)
y = np.zeros(N)

for i in range(1, N):
    y[i] = 0.7 * y[i - 1] + x[i]

plt.figure(figsize=(12, 6))

plt.plot(n, x, label='Input Signal x[n]', color='blue')
plt.plot(n, y, label='System Output y[n]', color='orange')

plt.title('LTI System Response: Input and Output Signals')
plt.xlabel('Samples')
plt.ylabel('Amplitude')
plt.grid()
plt.legend(loc='upper right')

plt.tight_layout()
plt.show()
```

**Problem 2: Non-Linear System Response**

```python
N = 100 
n = np.arange(N)  

x = np.sin(0.2 * np.pi * n)
y = x**2

plt.figure(figsize=(12, 8))

plt.subplot(2, 1, 1)
plt.scatter(n, x, label='Input Signal x[n]', color='blue', s=20)
plt.title('Non-Linear System Response')
plt.xlabel('Samples')
plt.ylabel('Amplitude')
plt.grid()
plt.legend(loc='upper right')

plt.subplot(2, 1, 2)
plt.bar(n, y, label='Output Signal y[n]', color='orange', width=0.5)
plt.xlabel('Samples')
plt.ylabel('Amplitude')
plt.grid()
plt.legend(loc='upper right')

plt.figure(figsize=(10, 5))
plt.plot(n, x, label='Input Signal x[n]', color='blue', linewidth=2)
plt.plot(n, y, label='Output Signal y[n]', color='orange', linewidth=2)
plt.title('Combined Input and Output Signals')
plt.xlabel('Samples')
plt.ylabel('Amplitude')
plt.grid()
plt.legend(loc='upper right')

plt.tight_layout()
plt.show()
```

**Problem 3: LTI System Response to a Noisy Discrete-Time Signal**

```python
N = 100  
n = np.arange(N)  

Xclean = np.sin(0.3 * np.pi * n)
noise = np.random.normal(0, 0.1, N) 
Xnoisy = Xclean + noise
Y = np.zeros(N)

for i in range(1, N):
    Y[i] = 0.5 * Y[i - 1] + 0.5 * Xnoisy[i]

plt.figure(figsize=(12, 12))

plt.subplot(3, 1, 1)
plt.plot(n, Xnoisy, label=r'Noisy Input Signal $X_{\text{noisy}}[n]$', color='red')
plt.title('LTI System Response to a Noisy Discrete-Time Signal')
plt.xlabel('Samples')
plt.ylabel('Amplitude')
plt.grid()
plt.legend(loc='upper right')

plt.subplot(3, 1, 2)
plt.plot(n, Y, label=r'Output Signal $Y[n]$', color='orange')
plt.xlabel('Samples')
plt.ylabel('Amplitude')
plt.grid()
plt.legend(loc='upper right')

plt.subplot(3, 1, 3)
plt.plot(n, Xclean, label=r'Clean Input Signal $X_{\text{clean}}[n]$', color='blue', linewidth=2)
plt.plot(n, Xnoisy, label=r'Noisy Input Signal $X_{\text{noisy}}[n]$', color='red', linewidth=2, alpha=0.7)
plt.plot(n, Y, label=r'Output Signal $Y[n]$', color='orange', linewidth=2)
plt.title('Combined Input and Output Signals')
plt.xlabel('Samples')
plt.ylabel('Amplitude')
plt.grid()
plt.legend(loc='upper right')

plt.tight_layout()
plt.show()
```
