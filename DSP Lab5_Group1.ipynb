```python
import numpy as np
import matplotlib.pyplot as plt
```

**Time-Invariant System (Multiplication by a Constant)**

```python
t = np.linspace(0, 10, 1000)
plt.figure(figsize = (8, 4))
tau = 2
plt.plot(t, 2*np.sin(t), label = r'$2\sin \theta$')
plt.plot(t, 2*np.sin(t-tau), label = r'$2\sin \theta -2$')
plt.title('Time-Invariant System: Multiplication by Constant')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```

**Time-Variant System (Multiplication by Time)**

```python
t = np.linspace(0, 10, 1000)
plt.figure(figsize = (8, 4))
tau = 2
plt.plot(t, t*np.sin(t), label = r'$2\sin \theta$')
plt.plot(t, (t-tau)*np.sin(t-tau), label = r'$2\sin \theta -2$')
plt.title('Time-Variant System: Multiplication by Time')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```

**Time-Invariant System (Differentiation)**

```python
t = np.linspace(0, 10, 1000)
plt.figure(figsize = (8, 4))
tau = 2
plt.plot(t, np.gradient(np.sin(t),t), label = r'Velocity $y(t) = \frac{d}{dt}\sin(t) $')
plt.plot(t, np.gradient(np.sin(t-tau),t), label = r'Shifted Velocity $y(t - \tau)$')
plt.title('Time-Invariant System: Differentiation')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```
